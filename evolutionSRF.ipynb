```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.figure(figsize=(12,6))

dates = ["2018-11-15", "2019-02-01", "2019-03-15", "2019-06-03", "2019-08-19", "2019-12-05", "2020-02-08", "2020-05-03"]
values = [0, 100, 120, 80, 2, 1, 0, 0]

dates2 = ["2018-11-15", "2019-02-01", "2019-03-15", "2019-06-03", "2019-08-19", "2019-12-05"]
dates3 = ["2019-12-05", "2020-02-08", "2020-05-03"]
values2 = [0, 130, 160, 60, 0, 0]
values3 = [0, 150, 130]

date_objects = [mdates.datestr2num(date) for date in dates]
date_objects2 = [mdates.datestr2num(date) for date in dates2]
date_objects3 = [mdates.datestr2num(date) for date in dates3]

plt.axvspan(*mdates.datestr2num(["2018-11-15", "2019-02-01"]), alpha=0.2, hatch='\\')

plt.plot(date_objects, values, 'b', label='Non-recurrent')
plt.plot(date_objects2, values2, 'r', label='Recurrent')
plt.plot(date_objects3, values3, 'r', linewidth=2)
plt.fill_between(date_objects3, values3, color='red', alpha=0.4)

plt.axvspan(*mdates.datestr2num(["2019-08-19", "2019-12-05"]), color='grey', alpha=0.3)
plt.axvline(*mdates.datestr2num(["2019-02-01"]), linewidth=2, color='black')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.gcf().autofmt_xdate()


date_labels = ["visit 1", "visit 2", "visit 3", "visit 4", "visit 5", "visit 6", "visit 7", "visit 8"]
plt.xticks(ticks=date_objects, labels=date_labels)

plt.legend(bbox_to_anchor = (1.17, 0.40), loc='center right')
plt.ylabel('Volume of SRF [nl]')
plt.xlabel("Time")
plt.title("Evolution of SRF for recurrent vs non-recurrent patient", fontweight="bold")

# Show the plot
plt.show()
```
